##  Problem Statement
Mental health issues are becoming increasingly prevalent among technology professionals due to high work pressure, lack of employer support, and poor work-life balance.  
This project aims to perform **exploratory data analysis (EDA)** to identify trends, risk factors, and correlations in the dataset. The findings can help organizations build better support systems for employee well-being.


```python
# Basic Libraries
import numpy as np
import pandas as pd

# Visualization
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset (replace path with your dataset file)
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Datasets/survey.csv")

# Preview data
df.head()
```

```python
# Dataset shape
print("Rows:", df.shape[0], "Columns:", df.shape[1])

# Data types and nulls
df.info()

# Summary statistics (numeric + categorical)
df.describe(include="all").T
```

```python
# Check missing values
df.isnull().sum()

# Drop irrelevant columns (if any, e.g., timestamps, IDs)
df = df.drop(columns=["Timestamp"], errors="ignore")

# Fill or drop missing values (example: drop rows with too many NaNs)
df = df.dropna(thresh=5)

# Standardize categorical values (e.g., Gender cleanup)
df["Gender"] = df["Gender"].str.strip().str.title()
```

```python
# Gender distribution
plt.figure(figsize=(6,4))
sns.countplot(x="Gender", data=df)
plt.title("Gender Distribution")

# Age distribution
plt.figure(figsize=(6,4))
sns.histplot(df["Age"], bins=20, kde=True)
plt.title("Age Distribution of Respondents")

# Country frequency
plt.figure(figsize=(8,5))
df["Country"].value_counts().head(10).plot(kind="bar")
plt.title("Top 10 Countries in Survey")
plt.ylabel("Count")
```

```python
# Gender vs Mental Health Condition
plt.figure(figsize=(6,4))
sns.countplot(x="Gender", hue="treatment", data=df)
plt.title("Gender vs Mental Health Treatment")

# Age vs Treatment
plt.figure(figsize=(6,4))
sns.boxplot(x="treatment", y="Age", data=df)
plt.title("Age Distribution vs Treatment")

# Company Size vs Treatment
plt.figure(figsize=(8,5))
sns.countplot(x="no_employees", hue="treatment", data=df)
plt.title("Company Size vs Treatment")
plt.xticks(rotation=45)
```

```python
# Employer Support vs Treatment
plt.figure(figsize=(7,4))
sns.countplot(x="benefits", hue="treatment", data=df)
plt.title("Employer Benefits vs Treatment")

# Remote Work vs Mental Health
plt.figure(figsize=(7,4))
sns.countplot(x="remote_work", hue="treatment", data=df)
plt.title("Remote Work vs Treatment")
```

```python
# Convert categorical to numeric for correlation (quick mapping)
df_num = df.copy()

for col in df_num.select_dtypes(include="object").columns:
    df_num[col] = df_num[col].astype("category").cat.codes

plt.figure(figsize=(10,6))
sns.heatmap(df_num.corr(), cmap="coolwarm", annot=False)
plt.title("Correlation Heatmap of Features")
```

#   Key Insights
1. Majority of respondents belong to the age group X–Y years.
2. Gender imbalance in reporting: Male vs Female mental health reports differ.
3. Remote work has [positive/negative] association with reported mental health issues.
4. Companies with benefits/support policies have lower reported treatment needs.
5. Age group < 25 has [higher/lower] tendency to report issues.


#  Conclusion
- This EDA highlights important workplace factors (company size, remote work, employer support) that influence mental health in tech professionals.  
- Employers with benefits/support structures show better employee mental well-being.
